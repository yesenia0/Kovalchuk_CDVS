# Завдання 2 

```python
import numpy as np
```

```python
# Створення діапазонів для x і y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
```

```python
# Задана функція
Z = np.sin(np.sqrt(X**2 + Y**2))
```

```python
# Побудова графіку поверхні
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title('Графік поверхні для z = sin(sqrt(x^2 + y^2))')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```
