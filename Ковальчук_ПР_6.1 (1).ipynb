```python
import altair as alt
from vega_datasets import data
```

```python
source = data.wheat()
```

```python
chart = alt.Chart(source).mark_bar().encode(
    x='year:O',
    y='wheat:Q',
    color=alt.condition(
        alt.datum.year == 1810,
        alt.value('orange'), 
        alt.value('blue')  
    )
).properties(
    width=600,
    height=400,
    title='Кількість пшениці за роками'
)
```

```python
chart
```

```python
#2
```

```python
average_rule = alt.Chart(source).mark_rule(color='red').encode(
    y='mean(wheat):Q',
    size=alt.value(2) 
)
```

```python
combined_chart = (chart + average_rule)
```

```python
combined_chart = combined_chart.properties(
    title='Кількість пшениці та середнє значення',
    config={'legendTitleFontSize': 14, 'legendLabelFontSize': 12}
)
```

```python
combined_chart
```

```python
combined_chart.save('wheat_production_chart.html')
```

```python
source = data.stocks()
```

```python
chart = alt.Chart(source).mark_line(point=True).encode(
    x='date:T',
    y='price:Q',
    color='symbol:N',  
    tooltip=['date:T', 'price:Q']  
).properties(
    width=800,
    height=400,
    title='Ціни акцій різних компаній з часом'
).configure_legend(
    title=None  
)
chart
```

```python
chart.save('stocks_chart.html')
```
