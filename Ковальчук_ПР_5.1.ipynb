```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import mplcursors
```

```python
#1
```

```python
iris = sns.load_dataset('iris')
```

```python
plt.figure(figsize=(10, 8))
scatter = plt.scatter(x='sepal_width', y='sepal_length', data=iris, 
                      c='petal_length', s=iris['petal_length']*10, cmap='viridis', alpha=0.7)
plt.colorbar(scatter, label='Petal Length')

def on_hover(event):
    index = event.ind[0]
    petal_width = iris.iloc[index]['petal_width']
    plt.gca().set_title(f'Petal Width: {petal_width}')

mplcursors.cursor(hover=True).connect("add", on_hover)

plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')
plt.title('Scatter Plot of Iris Dataset')
plt.show()
```

```python
data = {
    'nation': ['USA', 'China', 'Ukraine', 'Germany', 'France'],
    'medal': ['Gold', 'Silver', 'Bronze', 'Gold', 'Silver'],
    'total_medals': [100, 90, 80, 70, 60]
}
```

```python
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 6))

marker_size = 10

scatter = sns.scatterplot(x='total_medals', y='nation', hue='medal', style='medal',
                          s=marker_size, data=df, palette='Set1', markers={'Gold': 's', 'Silver': 'o', 'Bronze': '^'})
scatter.legend(title='Medal Type')
plt.xlabel('Total Medals')
plt.ylabel('Nation')
plt.title('Scatter Plot of Medal Counts by Nation and Type')
plt.show()
```

```python
import statsmodels.api as sm
```

```python
tips = sns.load_dataset("tips")
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x="total_bill", y="tip", data=tips, scatter_kws={'s': 50, 'alpha':0.6})

plt.title('Scatter Plot with Regression Line')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

X = sm.add_constant(tips['total_bill'])  
model = sm.OLS(tips['tip'], X).fit()
abline_values = [model.params[0] + model.params[1] * i for i in tips['total_bill']]
plt.plot(tips['total_bill'], abline_values, 'r', linestyle='dashed', linewidth=2)

plt.show()
```

```python
data = {
    'day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
    'tip': [15.5, 20.2, 18.0, 14.8, 22.5]
}
df = pd.DataFrame(data)
```
