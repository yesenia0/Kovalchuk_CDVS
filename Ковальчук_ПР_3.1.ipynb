```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1
```

```python
file_path = "C:\\Users\\E L I T E B O O K\\Downloads\\spotify-2023.csv"
```

```python
data = pd.read_csv(file_path, encoding='latin-1')
```

```python
top_10_songs = data.sort_values(by="streams", ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(top_10_songs["track_name"], top_10_songs["streams"])
plt.xlabel("Кількість прослуховувань")
plt.title("10 найпопулярніших пісень на Spotify у 2023 році")
plt.gca().invert_yaxis()
```

```python
plt.show()
```

```python
# Так як дані про жанри пісень відсутні і завдання зробити неможливо. Я зробила аналог на основі ключів до пісень
```

```python
top_10_songs = data.sort_values(by="streams", ascending=False).head(10)
```

```python
key_counts = top_10_songs["key"].value_counts()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(key_counts, labels=key_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Розподіл ключів серед найпопулярніших пісень у 2023 році")
```

```python
plt.show()
```

```python
artist_popularity = data.groupby("artist(s)_name")["streams"].sum().reset_index()
```

```python
top_10_artists = artist_popularity.sort_values(by="streams", ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_10_artists["artist(s)_name"], top_10_artists["streams"])
plt.xlabel("Артист")
plt.ylabel("Кількість прослуховувань")
plt.title("10 найпопулярніших артистів на Spotify у 2023 році")
plt.xticks(rotation=45)
```

```python
plt.show()
```

```python
track1_name = "Die For You"
track2_name = "I Wanna Be Yours"
track3_name = "Another Love"
```

```python
track1_data = data[data["track_name"] == track1_name]
track2_data = data[data["track_name"] == track2_name]
track3_data = data[data["track_name"] == track3_name]
```

```python
track_names = [track1_name, track2_name, track3_name]
track_listens = [track1_data["streams"].values[0], track2_data["streams"].values[0], track3_data["streams"].values[0]]
```

```python
plt.figure(figsize=(10, 6))
plt.barh(track_names, track_listens, color=['red', 'blue', 'green'])
plt.xlabel("Кількість прослуховувань")
plt.title("Порівняння кількості прослуховувань трьох треків")
plt.gca().invert_yaxis() 
```

```python
plt.show()
```
