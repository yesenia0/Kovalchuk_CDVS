```python
male_literacy_rate = df['Youth_15_24_Literacy_Rate_Male']
female_literacy_rate = df['Youth_15_24_Literacy_Rate_Female']
countries = df['Countries and areas']
```

```python
bar_width = 0.4
bar_positions = np.arange(len(countries))

plt.barh(bar_positions, male_literacy_rate, height=bar_width, label='Male Literacy Rate', color='blue')
plt.barh(bar_positions + bar_width, female_literacy_rate, height=bar_width, label='Female Literacy Rate', color='pink')

plt.yticks(bar_positions + bar_width / 2, countries)  
plt.xlabel('Literacy Rate')
plt.title('Порівняння рейтингів грамотності чоловіків та жінок')

plt.legend()
plt.show()
```
