```python
data_2015 = data[data['date'].str.contains('2015')]
```

```python
# Розділення стовпця "date" на рік та місяць
data_2015['year'] = pd.to_datetime(data_2015['date']).dt.year
data_2015['month'] = pd.to_datetime(data_2015['date']).dt.month
```

```python
# Групування даних за місяцем та обчислення загальної кількості відвідувачів
total_visitors_by_month = data_2015.groupby('month')['avila_adobe', 'firehouse_museum', 'chinese_american_museum', 'america_tropical_interpretive_center'].sum()
```

```python
# Знаходження місяця з найвищою і найнижчою загальною кількістю відвідувачів
highest_month = total_visitors_by_month.sum(axis=1).idxmax()
lowest_month = total_visitors_by_month.sum(axis=1).idxmin()
```

```python
# Визначення назв місяців
month_names = {
    1: 'Січень',
    2: 'Лютий',
    3: 'Березень',
    4: 'Квітень',
    5: 'Травень',
    6: 'Червень',
    7: 'Липень',
    8: 'Серпень',
    9: 'Вересень',
    10: 'Жовтень',
    11: 'Листопад',
    12: 'Грудень'
}

highest_month_name = month_names[highest_month]
lowest_month_name = month_names[lowest_month]
```

```python
print("Місяць з найвищою загальною кількістю відвідувачів для 2015 року:", highest_month_name)
print("Місяць з найнижчою загальною кількістю відвідувачів для 2015 року:", lowest_month_name)
```

```python
data_2018['year'] = pd.to_datetime(data_2018['date']).dt.year
data_2018['month'] = pd.to_datetime(data_2018['date']).dt.month
```

```python
# Визначення літніх та зимових місяців
summer_months = [6, 7, 8]
winter_months = [12, 1, 2]
```

```python
# Фільтрація даних для літніх і зимових місяців 2018 року
summer_visitors = data_2018[(data_2018['month'].isin(summer_months)) & (data_2018['avila_adobe'].notnull())]
winter_visitors = data_2018[(data_2018['month'].isin(winter_months)) & (data_2018['avila_adobe'].notnull())]
```

```python
# Обчислення кількості відвідувачів для літніх і зимових місяців
summer_visitors_count = summer_visitors['avila_adobe'].sum()
winter_visitors_count = winter_visitors['avila_adobe'].sum()
```

```python
print("Кількість відвідувачів музею 'Avila Adobe' у літні місяці (2018 рік):", summer_visitors_count)
print("Кількість відвідувачів музею 'Avila Adobe' у зимові місяці (2018 рік):", winter_visitors_count)
```

```python
data['year'] = pd.to_datetime(data['date']).dt.year
data['month'] = pd.to_datetime(data['date']).dt.month
```

```python
# Фільтрація даних за 2017 рік
data_2017 = data[data['year'] == 2017]
```

```python
# Вибірка стовпців з кількістю відвідувачів у кожному музеї
selected_columns = ['avila_adobe', 'firehouse_museum', 'chinese_american_museum', 'america_tropical_interpretive_center']
```

```python
# Групування даних за місяцем та обчислення середньої відвідуваності кожного музею
monthly_visitors = data_2017.groupby('month')[selected_columns].mean()
```

```python
# Побудова графіка ліній
plt.figure(figsize=(10, 6))
plt.plot(monthly_visitors.index, monthly_visitors['avila_adobe'], label='Avila Adobe', marker='o')
plt.plot(monthly_visitors.index, monthly_visitors['firehouse_museum'], label='Firehouse Museum', marker='o')
plt.plot(monthly_visitors.index, monthly_visitors['chinese_american_museum'], label='Chinese American Museum', marker='o')
plt.plot(monthly_visitors.index, monthly_visitors['america_tropical_interpretive_center'], label='America Tropical Interpretive Center', marker='o')

# Налаштування графіка
plt.xlabel('Місяць')
plt.ylabel('Середня відвідуваність')
plt.title('Відвідуваність музеїв протягом 2017 року')
plt.xticks(monthly_visitors.index, ['Січень', 'Лютий', 'Березень', 'Квітень', 'Травень', 'Червень', 'Липень', 'Серпень', 'Вересень', 'Жовтень', 'Листопад', 'Грудень'], rotation=45)
plt.legend()
plt.grid(True)
```

```python
# Роки для аналізу
years = [2014, 2015, 2016, 2017]
```

```python
# Побудова графіків гістограми
plt.figure(figsize=(12, 10))

for i, year in enumerate(years):
    data_year = data[data['year'] == year]
    
    plt.subplot(2, 2, i+1)
    for j, museum in enumerate(selected_columns):
        color = ['blue', 'green', 'yellow', 'gray'][j] 
        plt.hist(data_year[data_year[museum].notnull()][museum], bins=12, color=color, alpha=0.6, label=museum)

    plt.title(f'Відвідуваність у {year} році')
    plt.xlabel('Місяць')
    plt.ylabel('Кількість відвідувачів')
    plt.xticks(range(1, 13), ['Січень', 'Лютий', 'Березень', 'Квітень', 'Травень', 'Червень', 'Липень', 'Серпень', 'Вересень', 'Жовтень', 'Листопад', 'Грудень'], rotation=45)
    plt.legend()
```
